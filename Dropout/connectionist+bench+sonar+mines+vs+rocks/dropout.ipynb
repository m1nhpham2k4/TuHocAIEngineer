```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./sonar_all.csv', header=None)
df.sample(5)
```

```python
df.shape
```

```python
df[60].value_counts()
```

```python
# Tách dữ liệu chuẩn hóa về X và y
X = df.drop(60, axis='columns')
y = df[60]
X
```

```python
y = pd.get_dummies(y, drop_first=True)
y # R -> 1; M -> 0  
```

```python
y.value_counts()
```

```python
# Chia dữ liệu tập train thành 80% và tập test thành 20%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# Fully Connected 60 hidden layers -> 30 hidden layers -> 15 hidden layers -> 1 hidden layers
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

# Optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Train
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

Nhận thấy accuracy ở tập train có độ chính xác 100% nhưng với tập test chỉ có khoảng 71,42% (Dấu hiện overfitting)

```python
# So sánh kết quả dự đoán với nhãn ban đầu
y_pred = model.predict(X_test).reshape(-1)

# Làm tròn prediction label
y_pred = np.round(y_pred)
print(y_pred[:10])
print(y_test[:10])
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
# Tinh chỉnh lại model dùng dropout loại bỏ bớt các hidden layers
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid'),
])

# Optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

Nhận thấy kết quả đã cải thiện hơn

=> Khi train model tránh bị overfitting chúng ta nên tinh chỉnh lại model xóa bớt các hidden layers phía trước nó để học mới hơn.
